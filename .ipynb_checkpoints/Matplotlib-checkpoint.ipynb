## Changing labels and axis limits

```python
plt.plot(x,y)
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim(0, np.pi*2)
plt.axis('equal') #aspect ratio 
plt.show()
```

```python
plt.scatter(x, y, s= 300, marker = 'd', color = 'maroon') #d for diamonds
plt.scatter(x,w, s = 200, marker = 'H', color = 'darkslategrey')
plt.xlim([0,.2])
plt.show()
```

```python
#arg linestyle = '-' equil to 'solid' or '--' is 'dashed' etc ':' is 'dotted' 
```

```python
np.random.seed(1)
x = np.random.random(100)
y = np.random.random(100)
plt.figure(figsize = (4,4))  # define the properties of the figure first
plt.scatter(x,y)             # then add content
plt.axis('equal')            # alter properties
plt.title('2-dimensional random uniform values')
plt.show()
```

```python
#point size and color options 
z = 100 * np.random.randint(1,11,100)  # use for size
w = np.random.randint(0,5,100) # use for color
print(z)
print(w)
```

```python
colors = np.array(['r', 'b', 'g', 'm', 'y'])
c = colors[w]
print(c)
```

```python
plt.figure(figsize = (8,8))
plt.scatter(x,y, s=z, c = c, alpha = 0.4)
plt.axis('equal')
plt.show()
```

```python
#marker could be '*' - can't use markers for a variable 
```

```python
# You can call 'scatter' multiple times to plot different groups, each with its own label
# then call plt.legend() to add a legend with the appropriate labels
plt.figure(figsize = (5,5))
plt.scatter(x[0:10] , y[0:10] , s = 100, c = 'red' , marker = "*", label = 'group A')  
plt.scatter(x[11:20], y[11:20], s = 100, c = 'blue', marker = "o", label = 'group B')
plt.legend()
plt.show()
```

```python
#Subplots 
plt.figure(figsize=(9,6))
for i in range (1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5, 'plot ' + str((2,3,i)), fontsize = 14, ha='center') #Not zero indexed like normal python 
```

```python
grid = plt.GridSpec(2,3,wspace = 0.4, hspace = 0.3) #0 indexed 
plt.subplot(grid[0, 1:3]) #the top row columns 1 to 3 
#subplots are not ideal, better to just plot individually 
```

## Histograms

```python
data = np.random.randn(1000)
plt.hist(data)
plt.show()
```

```python
#opt args are bins = num, alpha = transparency, 0 to 1, color = fill color and edgecolor 
#histtype = stepfilled won't draw the lines between bars 
```

```python
x1 = np.random.normal(0, 0.8, 10000)
x2 = np.random.normal(-2, 1, 10000)
x3 = np.random.normal(3, 2, 10000)

# set up a dictionary with arguments
kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=20, edgecolor = 'black')

# use the same arguments for all of the histograms without the need to copy and paste
plt.hist(x1, **kwargs, color = 'navy')
plt.hist(x2, **kwargs, color = "purple")
plt.hist(x3, **kwargs)
plt.show()
```

## Two dimensional functions 

```python
#use meshgrid 
x = np.array([0,1,2,3])
y = np.array([-1,0,1, 2])
print(x)
print(y)
X, Y = np.meshgrid(x,y) # sees x has length 3, y has length 4, so the result is a 4 x 3 grid
```

```python
def f(x, y):
    return x * y
X, Y = np.meshgrid(x,y)
Z = f(X, Y)
print(Z)
```

```python
x = np.linspace(-3, 3, 101)
y = np.linspace(-3, 3, 101)
X, Y = np.meshgrid(x, y)

def g (x, y): 
    return np.exp(-(x**2)) * np.exp(-(y**2)) # product of 2 normal distributions

Z = g(X, Y)
```

```python
Z.shape
```

```python
plt.contour(X, Y, Z)
plt.axis('equal')
plt.show()
```

## Plotting from a Pandas Series

```python
ts = pd.Series(np.random.randn(1000),
   index=pd.date_range('1/1/2000', periods=1000))
```

```python
ts = ts.cumsum()
```

```python
ts.plot() #will try to automatically add labels 
```

## Plotting From a Data Frame

```python
df = pd.DataFrame(np.random.randn(1000, 4), index = ts.index, columns = list('ABCD'))
```

```python
df = df.cumsum()
plt.figure()
df.plot()
```
